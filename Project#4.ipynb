<b>Q#5</b> Consider the <b><i>Team</i></b> file. This file contains a winning teams. The first line in the file is the name of the team that won in 1903, and the last line is the name of the team that won in 2006. Write a program that reads this file and creates a dictionary in which the keys are the names of the teams, and each key’s associated value is the number of times the team has won. The program should also create a dictionary in which the keys are the years, and each key’s associated value is the name of the team that won that year.
The program should prompt the user for a year in the range of 1903 through 2007. It should then display the name of the team that won that year, and the number of times that team has won.
<br><br>
<b><i>Example:</i></b><br/>
Enter a year in the range 1903-2006: 2000
<br>
The team that won in 2000 is the Anaheim Angels.
<br>
They won  1 times.

```python
def read_file():
    
    program11file = open('Team.txt', 'r')

    
    winners = {}
    years = {}

    year = 1903
    
    for name in program11file.readlines():
        team_name = name.rstrip()
        if team_name in winners.keys():
            winners[team_name] += 1
        else:
            winners[team_name] = 1

        years[year] = team_name
        year += 1

    
    return winners, years


def display_result(year, winners, years):
    
    
    if year < 1903 or year > 2006:
        print('\nThe data for the year', year, 'is not included in our database.')
    else:
    
    
        team_name = years[year]
        print('\nTeam won in', year, 'is', team_name)
        print(team_name, 'won', winners[team_name], 'times.')


def assign_year():
    
    year = int(input('Enter year between 1903-2006 or 0 to quit: '))
    
    return year


def main():
    winners, years = read_file()

    year = assign_year()

    
    while year != 0:
        display_result(year, winners, years)
        print()
        
        year = assign_year()


main()
```
